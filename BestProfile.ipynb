```python
profitList = []
```

```python
for index, row in df.iterrows():
    total = float(row['Price']) * float( row['Volume']) * float(row['Profit margin'])
    profitList.append({'Product Type':row['Product Type'],'Product #':row['Product #'],'Profit':total})
```

```python
print(profitList)
```

```python
bestProfit =0
bestProfitName=''

worstProfit=0
worstProfitName=''
for listInfo in profitList:
    if bestProfit < listInfo['Profit']:
        bestProfit = listInfo['Profit']
        bestProfitName = listInfo['Product Type'] + '->' + str(listInfo['Product #'])
    
    if worstProfit ==0 or worstProfit > listInfo['Profit']:
        worstProfit = listInfo['Profit']
        worstProfitName = listInfo['Product Type'] + '->' + str(listInfo['Product #'])

print("Producto con mayor rendimiento")
print(bestProfit)
print(bestProfitName)

print('Peor rendimiento')
print(worstProfit)
print(worstProfitName)
```
